```python
from sar_imaging_pytorch.imaging.lia import LIA
from sar_imaging_pytorch.imaging.mf import MF
from sar_imaging_pytorch.imaging.rma import RMA

import torch
```

```python
lia = LIA(z0=1000, dx=2, dy=2, im_size=[-100, 100, -80, 80], kk=4700)

# Prepare your SAR data (assuming it's a 2D tensor)
sar_data = torch.randn(100, 100)  # Replace with your actual SAR data

# Process the data
result = lia(sar_data)

# Optionally, plot the result
lia.plot_result(result)
```

```python
# Initialize the Matched Filtering algorithm
mf = MF(x_step_m=2, y_step_m=2, xy_size_t=200)  # or xy_size_t=[x0, x1, y0, y1] for bounding box

# Prepare your SAR data and matched filter (assuming they're 2D tensors)
sar_data = torch.randn(100, 100)  # Replace with your actual SAR data
matched_filter = torch.randn(100, 100)  # Replace with your actual matched filter

# Process the data
result = mf(sar_data, matched_filter)

# Optionally, plot the result
mf.plot_result(result)
```

```python
rma = RMA(nFFTspace=1024, z0=1000, dx=2, dy=2, bbox=200)  # or bbox=[x0, x1, y0, y1] for bounding box

# Prepare your SAR data (assuming it's a 2D tensor)
sar_data = torch.randn(100, 100)  # Replace with your actual SAR data

# Process the data
result = rma(sar_data)

# Optionally, plot the result
rma.plot_result(result)
```
